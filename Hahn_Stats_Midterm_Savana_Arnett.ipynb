## Part 2: Linear Fit

Follow the steps in the [Linear Fit notebook]() to run a linear fit on your data.

```python
x = diameters
y = circumferences
```

```python
#The x and y have to be difined, so we set x = diameters and y = circumferences from the data.
```

```python
best_loss = 100000000

best_m = 0
best_b = 0

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        print("Now Testing: ", m_guess, b_guess)

        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:

            best_m = m_guess
            best_b = b_guess
            best_loss = loss
```

```python
#This lists all of the numbers the program thinks could be best fit.
```

Plot the line of best fit:

Use blue dots for the collected data and a red line for the estimate.

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r');
plt.plot(diameters,circumferences,'b.');
plt.show()

print(best_loss,best_m,best_b)
```
